```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
ImageId = test_data.index + 1
y=train_data.iloc[:,0]
X=train_data.iloc[::,1:]
X_test=test_data
```

```python
y
```

```python
X
```

```python
X_test
```

```python
X=X/255
X_test=X_test/255
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.1,random_state=1)
X_train.shape,X_val.shape,y_train.shape,y_val.shape,X_test.shape
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=50, alpha=1e-5, solver='sgd', verbose=10, random_state=1, learning_rate='constant', learning_rate_init=0.1).fit(X_train, y_train)
print(model.score(X_train, y_train))
```

```python
y_pred=model.predict(X_val)
print(model.score(X_val,y_val))
```

```python
df_val=pd.DataFrame({'actual_value':y_val, 'predicted_value':y_pred})
df_val.reset_index(drop=True, inplace=True)
df_val
```

```python
pred=model.predict(X_test)
pred
```

```python
output = pd.DataFrame({'ImageId': ImageId, 'Label': pred})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
